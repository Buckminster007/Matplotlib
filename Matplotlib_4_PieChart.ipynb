```python
import matplotlib.pyplot as plt
```

```python
#Basic pie chart
sizes=[3, 4, 6, 1]
plt.pie(sizes)
plt.show()
```

```python
#add labels to the pies and customize colours
#adding a title to the pie chart
sizes=[3, 4, 6, 1]
labels=['A','B','C','D']
explode=[0.1, 0, 0, 0]
colors=['red','black','orange','yellow']
plt.pie(sizes, colors=colors, explode=explode, labels=labels)
#explode feature helps one or more pies to come out of the graph. This helps in presentation
plt.title('split among alphabets')
plt.show()
```

```python
sizes=[3, 4, 6, 1]
labels=['A','B','C','D']
colors=['red','blue','green','yellow']
explode=[0.1, 0, 0, 0]
plt.pie(sizes, colors=colors, labels=labels, explode=explode, wedgeprops={'edgecolor':'black'}, autopct='%.2f%%')
#wedgeprop->edgecoloe added to make the edges/borders black. It beautifies the graph
#autopct is used to represent the percentage of area among the pies.
#%.2f% means 2 digits after decimal. %.2%% means 2 digits after decimal followed by a % sign 
plt.title('split among alphabets')
plt.axis("equal") 

plt.show()
```

```python
sizes=[3, 4, 6, 1]
labels=['A','B','C','D']
colors=['red','blue','green','yellow']
explode=[0.1, 0, 0, 0]
plt.style.use("fivethirtyeight")
plt.pie(sizes, colors=colors, labels=labels, explode=explode, wedgeprops={'edgecolor':'black'}, autopct='%.2f%%', counterclock=False, startangle=90, shadow=True)
#counterclock turned to false to make the chart clockwise. 
#startangle helps the first pie start from a particular angle from the +ve x-axis.
plt.title('split among alphabets')
plt.axis("equal")#to make the chart look circular--initially it might be little oval 

plt.show()
```
