```python
b=[1,2,1,3,5,4,5,4,3,3,20,7,6,7,8,8,20,22,11,14]
```

```python
plt.hist(b)
plt.show()
```

So it is merging values in larger bins because 1 hasen't obviously occured 6 times.
And we can't understand the size of the bin which leads us to being unable to understand which element occcured how frequently


```python
import numpy as np
xt=np.arange(23)
```

```python
xt
```

```python
#we will use edgecolor and xticks to make the histogram more understandable
plt.hist(b, edgecolor='black')
plt.xticks(xt)
plt.show()
```

We can also set the number of bins ourselves
as the highest number in the list b was 22. 
So we have 21 bins if we want one bin for each. 
Bin 1 is [1,2). It will represent the frequency of 1

```python
plt.hist(b,bins=21, edgecolor='black')
plt.xticks(xt)
plt.show()
```

Suppose we want to alter the size of bins according to our need


```python
plt.hist(b, bins=[1,5,9,14,17,20],edgecolor='black')
plt.xticks([1,5,9,14,17,20])
plt.show()
```

```python
import pandas as pd
data=pd.read_csv('histogram_data.csv')
data
```

```python
age=data['Age']
```

```python
age
```

```python
plt.hist(age,bins=list(np.arange(10,101,10)), edgecolor='black')

plt.title('Number of Users by age group')
plt.xlabel('Age')
plt.ylabel('Number Of Users')

plt.style.use('fivethirtyeight')
plt.grid(True)

plt.xticks(list(np.arange(10,81,10)))
plt.yticks(list(np.arange(5000,50000,5000)))

plt.savefig('plot2.png')
plt.show()
```

Well, there are some users in 60-70 and 70-80 agroup. 
But those get dwarfed by the no of users in 20-30 or 30-40
So to highlight those low numbers, we use **log** argument

```python
plt.hist(age,bins=list(np.arange(10,101,10)), edgecolor='black', log=True)

plt.title('Number of Users by age group')
plt.xlabel('Age')
plt.ylabel('Number Of Users')

plt.style.use('fivethirtyeight')
plt.grid(True)

plt.xticks(list(np.arange(10,101,10)))
#plt.yticks(list(np.arange(5000,50000,5000)))

plt.savefig('plot3.png')
plt.show()
```

Suppose now you want to have a line on the Graph. To represent e.g. Median Age or Mean Age. So we can use 
*plt.axvline()* function.

```python
plt.hist(age,bins=list(np.arange(10,101,10)), edgecolor='black')

plt.title('Number of Users by age group')
plt.xlabel('Age')
plt.ylabel('Number Of Users')

plt.style.use('fivethirtyeight')
plt.grid(True)

plt.xticks(list(np.arange(10,101,10)))
plt.yticks(list(np.arange(5000,50000,5000)))

median_age=np.median(age)
plt.axvline(median_age, color='orange', label='median age')
plt.legend()

plt.savefig('plot4.png')
plt.show()
```
